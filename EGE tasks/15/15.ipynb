```python
def f(a, x):
    return (a%9==0)and((280%x==0)<=((a%x!=0)<=(730%x!=0)))
k=0
for a in range(1, 1001):
    f=0
    for x in range(1,1000):
        if f(a,x)==0:
            f=1
            break
    if f==0:
        k+=1
print(k)
```

```python
def fu(x, y, a):
    return (x-2*y<3*a)or(2*y>x)or(3*x>50)
for a in range(1, 1000):
    f=0
    for x in range(1, 1000):
        if f==1:
            break
        for y in range(1, 1000):
            if fu(x, y, a)==0:
                f=1
                break
    if f==0:
        print(a)
        break
```
