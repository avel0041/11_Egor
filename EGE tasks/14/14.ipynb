```python
def f(x):
    s=''
    while x!=0:
        s+=str(x%4)
        x=x//4
    s=s[::-1]
    return s
k=0
for y in range(0, 32):
    a=32**4+7*32**3+9*32**2+y*32+9 + 7*128**3+y*128**2+9*128+3
    if f(a).count('0')>=k:
        k=f(a).count('0')
        A_max = f(a)      
print(sum(map(int, str(A_max))))
```
